```python
%sql
USE CATALOG dm_ws_7405611305977123;
```

```python
%sql
CREATE SCHEMA IF NOT EXISTS dm_ws_7405611305977123.NEW;
```

```python
%sql
CREATE TABLE IF NOT EXISTS NEW.TABLE_INFO(
  tid INT,
  table_name VARCHAR(100),
  container_name VARCHAR(100),
  path VARCHAR(100),
  filename VARCHAR(100)
);
```

```python
%sql
TRUNCATE TABLE NEW.TABLE_INFO;
INSERT INTO NEW.TABLE_INFO VALUES
(1,'transaction','container','landing/mysql_source','transaction.csv'),
(2,'orders','container','landing/mysql_source','orders.csv');

SELECT * FROM NEW.TABLE_INFO;
```

```python
%sql
DROP TABLE IF EXISTS NEW.CRED_INFO;

CREATE TABLE IF NOT EXISTS NEW.CRED_INFO(
  cid INT,
  tid INT,
  server_name VARCHAR(100),
  port_no VARCHAR(100),
  db_name VARCHAR(100),
  username VARCHAR(100),
  password VARCHAR(100)
)
```

```python
%sql
INSERT INTO NEW.CRED_INFO VALUES
(101,1,'localhost','3306','ABC','root','password'),
(102,2,'localhost','3306','ABC','root','password');

SELECT * FROM NEW.CRED_INFO;
```

```python
%sql
CREATE OR REPLACE VIEW NEW.joined_metadata AS
SELECT c.tid, table_name, container_name, path, filename, server_name, port_no, db_name, username, password
FROM NEW.TABLE_INFO AS t INNER JOIN NEW.CRED_INFO as c
ON c.tid = t.tid; 

SELECT * FROM joined_metadata;
```
